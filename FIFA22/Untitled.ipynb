```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from tqdm import tqdm
import numpy as np
import time
import time
from tqdm import tqdm
from itertools import chain
```

```python
URL = 'https://www.fifplay.com/fifa-22/players/?page={}'
```

```python
data = []
for page_num in tqdm(range(1,123)):
    current_page=URL.format(page_num)
    page = requests.get(current_page)
    soup = BeautifulSoup(page.content, "html.parser")
    table = soup.find('table', attrs={'class':'table listing table-vert-align table-nopadding table-hover table-striped'})
    table_body = table.find('tbody')

    rows = table_body.find_all('tr')
    for row in rows:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        data.append([ele for ele in cols if ele])
```

```python
data=pd.DataFrame(data)
data
```

```python
data.to_excel('FIFA_22.xlsx', index=False, header=True)
```
