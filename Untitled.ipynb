```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from tqdm import tqdm
import numpy as np
```

```python
URL = 'https://www.fifacm.com/players?page={}'
```

```python
#color_dict = {'&player_rating=75-80':'light-green',53
#              '&player_rating=81-99':'green',13
 #           '&player_rating=61-74':'yellow',439
 #          '&player_rating=40-60':'red'}130
```

```python
#tags
pl_name = 'player-name' # in 'a' tag
pl_info = 'player-position-cln'
pl_ovr1 = 'player-overall rating-search cm21 fifa-green-b'
pl_ovr2 = 'player-overall rating-search cm21 fifa-light-green-b'
pl_ovr3 = 'player-overall rating-search cm21 fifa-red-b'
pl_ovr4 = 'player-overall rating-search cm21 fifa-yellow-b'
pl_pot1 = 'player-potential rating-search cm21 fifa-green-b'
pl_pot2 = 'player-potential rating-search cm21 fifa-light-green-b'
pl_pot3 = 'player-potential rating-search cm21 fifa-red-b'
pl_pot4 = 'player-potential rating-search cm21 fifa-yellow-b'
pl_age = 'player-age pt-3'
pl_wage = 'player-value-wage pt-2'
pl_foot = 'player-wr pt-3'
pl_sf = 'player-skills pt-3'
pl_wf = 'player-wf pt-3'
pl_rc = 'player-igs pt-3'
pl_hei = 'player-height'
pl_wei = 'player-weight pl-1'
```

```python

name_data = []
ref_data = []
info_data = []
ovr_data = []
pot_data = []
age_data = []
wage_data = []
foot_data = []
work_data = []
sf_data = []
wf_data = []
rc_data = []
hei_data = []
wei_data = []
club_data=[]
country_data=[]

for page_num in tqdm(range(1,634)):
    current_page=URL.format(page_num)
    page = requests.get(current_page)
    soup = BeautifulSoup(page.content, "html.parser")
    
    player_names = soup.findAll('div',{'class':pl_name})
    for player_name in player_names:
        name_data.append(player_name.find('a').text)
        ref_data.append((player_name.find('a',href=True)['href']))
    
    player_infos = soup.findAll('div',{'class':pl_info})
    for player_info in player_infos:
        info_data.append(player_info.text)
    
    player_infos = soup.findAll('div',{'class':[pl_ovr1,pl_ovr2,pl_ovr3,pl_ovr4]})
    for player_info in player_infos:
        ovr_data.append(player_info.text)
    
    player_infos = soup.findAll('div',{'class':[pl_pot1,pl_pot2,pl_pot3,pl_pot4]})
    for player_info in player_infos:
        pot_data.append(player_info.text)
    
    player_infos = soup.findAll('div',{'class':pl_age})
    for player_info in player_infos:
        age_data.append(player_info.text)
    
    player_infos = soup.findAll('div',{'class':pl_wage})
    for player_info in player_infos:
        wage_data.append(player_info.text)
    
    player_infos = soup.findAll('div',{'class':pl_foot})
    i=0
    for player_info in player_infos:
        if i%2==0:
            work_data.append(player_info.text)
        else:
            foot_data.append(player_info.text)
        i+=1
    
    player_infos = soup.findAll('div',{'class':pl_sf})
    for player_info in player_infos:
        sf_data.append(player_info.text)
    
    player_infos = soup.findAll('div',{'class':pl_wf})
    for player_info in player_infos:
        wf_data.append(player_info.text)
    
    player_infos = soup.findAll('div',{'class':pl_rc})
    i=0
    for player_info in player_infos:
        if i%2==0:
            rc_data.append(player_info.text)
        i+=1
    for a in soup.find_all('div',{'class':pl_hei}):
        hei_data.append(a.text)
    
    for a in soup.find_all('div',{'class':pl_wei}):
        wei_data.append(a.text)
    
    club_country=[]
    player_names = soup.findAll('img')
    for player_name in player_names:
        if 'title' in str(player_name):
            club_country.append(str(player_name)[str(player_name).find('title')+7:-3])
    i=0
    for player_info in club_country:
        if i%2==0:
            club_data.append(player_info)
        else:
            country_data.append(player_info)
        i+=1
```

```python
DATA=pd.DataFrame({'player_name':name_data, 'info':info_data ,'overall_rating':ovr_data, 'potential_rating':pot_data,
                   'age':age_data,'wages':wage_data,'primary_foot':foot_data,'work_rate':work_data,
                   'strong_foot':sf_data,'weak_foot':wf_data,'release_clause':rc_data,'ref_url':ref_data,'hieght':hei_data,
                   'weight':wei_data,'club':club_data,'country':country_data})
```

```python
DATA1=pd.read_excel('FIFA_Players.xlsx',engine='openpyxl')
```

```python
DATA1['ref_url'] = DATA1['ref_url'].apply(lambda x: 'https://www.fifacm.com'+str(x))
DATA1['positions']=DATA1['info'].apply(lambda x: str(x).split("|")[0].strip())
DATA1['contract']=DATA1['info'].apply(lambda x: str(x).split("|")[1].strip())
DATA1['transfer_value']=DATA1['wages'].apply(lambda x: str(x).split(" ")[2])
DATA1['salary']=DATA1['wages'].apply(lambda x: str(x).split(" ")[3])
DATA1['attacking_work_rate'] = DATA1['work_rate'].apply(lambda x: str(x).strip().split("/")[0])
DATA1['defensive_work_rate'] = DATA1['work_rate'].apply(lambda x: str(x).strip().split("/")[1])
DATA1['height(cm)'] = DATA1['hieght'].apply(lambda x: str(x).strip().split("|")[0])
DATA1['height(inch)'] = DATA1['hieght'].apply(lambda x: str(x).strip().split("|")[1])
DATA1['personality'] = DATA1['weight'].apply(lambda x: str(x).split("(")[0].lstrip().rstrip())
DATA1['weight(kg)'] = DATA1['weight'].apply(lambda x: str(x).split("(")[1].split("|")[0].lstrip().rstrip())
DATA1['weight(lbs)'] = DATA1['weight'].apply(lambda x: str(x).split("(")[1].split("|")[1].lstrip().rstrip().replace(")",""))
DATA1['contract_start_year'] = DATA1['contract'].apply(lambda x: str(x).strip().split("-")[0])
DATA1['contract_end_year'] = DATA1['contract'].apply(lambda x: str(x).strip().split("-")[1])
DATA1['contract_length'] = DATA1['contract_end_year'].astype(np.int) - DATA1['contract_start_year'].astype(np.int)
DATA1['contract_remaining'] = DATA1['contract_end_year'].astype(np.int)- 2021
DATA1['contract_remaining'] = DATA1['contract_remaining'].apply(lambda x: 0 if x<0 else x)
DATA1['contract_length'] = DATA1['contract_length'].apply(lambda x: np.nan if x<0 else x)
DATA1['position'] = DATA1['positions'].apply(lambda x: str(x).split(",")[0])
DATA1['other_positions'] = DATA1['positions'].apply(lambda x: str(x).split(",")[1:] if len(str(x).split(","))>1 else np.nan)
```

```python
DATA1.drop(columns=['info',
       'wages', 'work_rate', 'hieght', 'weight','positions'],inplace=True)
```

```python
DATA1.to_excel('Cleaned_FIFA_players.xlsx',index=False,header=True)
```
